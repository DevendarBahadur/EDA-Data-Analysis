# Data cleaning

```python
df['Genre'] = df['Genre'].apply(lambda x:x.replace('\n',''))
```

```python
df['Duration in min'] = df['Duration in min'].apply(lambda x:x.replace('min',''))
```

```python
df['Year of Release'] = df['Year of Release'].apply(lambda x:x.replace(')\n',''))
```

```python
df['Title'] = df['Title'].apply(lambda x:x.replace('\n',''))
```

```python
df['Votes in Views'] = df['Votes in Views'].apply(lambda x:x.replace(',',''))
```
