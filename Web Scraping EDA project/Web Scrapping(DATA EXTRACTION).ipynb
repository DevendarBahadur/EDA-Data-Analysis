```python
import numpy as np
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup

import matplotlib.pyplot as plt
import seaborn as sns

import time

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'

# importing everything what we need.
```

[Imdb](https://www.imdb.com/list/ls006266261/?sort=list_order,asc&st_dt=&mode=detail&page=1)

```python
Movie_name = []
Movie_year = []
duration = []
genres = []
rating=[]

for i in range(1,11):
    URL = 'https://www.imdb.com/list/ls006266261/?sort=list_order,asc&st_dt=&mode=detail&page={}'. format(i)
    
    page = requests.get(URL)
    htmlCode = page.text
    
    soup = BeautifulSoup(htmlCode)
    
    for x in soup.find_all('div', attrs={'class' : 'lister-item-content'}):

        mov = x.find('h3', attrs={'class' : 'lister-item-header'})
        if mov is None:
            Movie_name.append(np.NaN)
        else:
            Movie_name.append(mov.text.split(('('))[0])
            
        year = x.find('h3', attrs={'class' : 'lister-item-header'})
        if year is None:
            Movie_year.append(np.NaN)
        else:
            Movie_year.append(mov.text.split(('('))[1])
            
        dur = x.find('span', attrs={'class': "runtime"})
        if dur is None:
            duration.append(np.NaN)
        else:
            duration.append(dur.text)
            
        gen = x.find('span', attrs={'class': "genre"})
        if gen is None:
            genres.append(np.NaN)
        else:
            genres.append(gen.text)
            
        rat = x.find('span', attrs={'class': "ipl-rating-star__rating"})
        if rat is None:
            rating.append(np.NaN)
        else:
            rating.append(rat.text)
       
```

```python
votes = []
gross = []


for i in range(1,11):
    URL = 'https://www.imdb.com/list/ls006266261/?sort=list_order,asc&st_dt=&mode=detail&page={}'. format(i)
    
    page = requests.get(URL)
    htmlCode = page.text
    for j in soup.find_all('div',attrs = {"class":"lister-item-content"}):
        k =j.find_all("span",attrs={"name":"nv"})
        if k is None:
            votes.append(np.NaN)        
        else:
            votes.append(k)
```

```python
v=[]
for i in votes:
    if i[0] is None:
        v.append(np.NaN)
    else:
        v.append(i[0].text)
```

```python
G=[]
for i in votes:    
    if len(i)<2:
        G.append(np.NaN)
    else:
        G.append(i[1].text)
```

```python
D = {'Title':Movie_name,
     'Year of Release':Movie_year,
    'Duration in min':duration,
    'Genre':genres,
    'Rating':rating,
    'Votes in Views':v,
    "Gross in Millions":G}
```

```python
df = pd.DataFrame(D)
df
```
